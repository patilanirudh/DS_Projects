```python
data['checking_balance'].replace('< 0 DM',0,inplace=True)
data['checking_balance'].replace('1 - 200 DM',100,inplace=True)

data['checking_balance'].replace('> 200 DM',201,inplace=True)
# #data['checking_balance'].replace(np.NaN,0,inplace=True)
```

```python
data['savings_balance'].replace('< 100 DM',np.random.randint(1,99),inplace=True)
data['savings_balance'].replace('100 - 500 DM',250,inplace=True)
data['savings_balance'].replace('500 - 1000 DM',750,inplace=True)
data['savings_balance'].replace('> 1000 DM',np.random.randint(1001,1500),inplace=True)
#data['savings_balance'].replace('unknown',np.random.randint(1,200),inplace=True)
```

```python
data['checking_balance']=data['checking_balance'].replace(to_replace='unknown',value=np.nan)
data['checking_balance']=data['checking_balance'].astype(float)
data['checking_balance'].replace(np.nan,data['checking_balance'].median(),inplace=True)
```

```python
data['savings_balance']=data['savings_balance'].replace(to_replace='unknown',value=np.nan)
data['savings_balance']=data['savings_balance'].astype(float)
data['savings_balance'].replace(np.nan,data['savings_balance'].median(),inplace=True)
```

```python
data['employment_duration'].replace('1 - 4 years',2,inplace=True)
data['employment_duration'].replace('> 7 years',8,inplace=True)
data['employment_duration'].replace('4 - 7 years',6,inplace=True)
data['employment_duration'].replace('< 1 year',0.6,inplace=True)
data['employment_duration'].replace('unemployed',0,inplace=True)

```

```python
dummies=pd.get_dummies(data[['credit_history','purpose','other_credit','housing','job','phone']],
                      columns=['credit_history','purpose','other_credit','housing','job','phone'])
```

```python
data.dtypes
```

```python
dummies.head()
```

```python
data=pd.concat([data,dummies],axis=1)
```

```python
columns=['credit_history','purpose','other_credit','housing','job','phone']
data.drop(columns,axis=1,inplace=True)
```

## Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
x=data.drop(['default'],axis=1)
y=data[['default']]
#split x and y data into 70:30 ratio
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
#fit linear model
reg_model=LogisticRegression(solver='liblinear',max_iter=5,n_jobs=4)
reg_model.fit(x_train,y_train)
```

```python
print(reg_model.score(x_train,y_train))
print(reg_model.score(x_test,y_test))
```

## DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
Dt=DecisionTreeClassifier(max_depth=3)
Dt.fit(x_train,y_train)
print(Dt.score(x_train,y_train))
print(Dt.score(x_test,y_test))
```

## BaggingClassifier

```python
from sklearn.ensemble import BaggingClassifier
Bc=BaggingClassifier(n_estimators=22,base_estimator=Dt,max_samples=20)
Bc.fit(x_train,y_train)
print(Bc.score(x_train,y_train))
print(Bc.score(x_test,y_test))
```

## AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier(n_estimators=2,random_state=1)
ad.fit(x_train,y_train)

print(ad.score(x_train,y_train))
print(ad.score(x_test,y_test))
```

## GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gc=GradientBoostingClassifier(n_estimators=7)
gc.fit(x_train,y_train)
print(gc.score(x_train,y_train))
print(gc.score(x_test,y_test))
```

## RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=400,random_state=1,max_samples=19)
rf.fit(x_train,y_train)
print(rf.score(x_train,y_train))
print(rf.score(x_test,y_test))
```
