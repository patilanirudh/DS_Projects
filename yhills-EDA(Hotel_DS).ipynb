```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q3-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
low_lead_time,up_lead_time=remove_outlier(data['lead_time'])
data['lead_time']=np.where(data['lead_time']>up_lead_time,up_lead_time,data['lead_time'])
data['lead_time']=np.where(data['lead_time']<low_lead_time,low_lead_time,data['lead_time'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
#linear model building
#independent variable and dependent variable
x=data.drop(['is_canceled'],axis=1)
y=data[['is_canceled']]
```

```python
#split x and y data into 70:30 ratio
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

## Logistic Regression

```python
#fit linear model
reg_model=LogisticRegression()
reg_model.fit(x_train,y_train)
```

```python
reg_model.score(x_train,y_train)
```

```python
reg_model.score(x_test,y_test)
```

```python
# pd.set_option("display.max_rows",None)
predictions=reg_model.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
cm=metrics.confusion_matrix(y_test,predictions,labels=[1,0])
df_cm=pd.DataFrame(cm,index=[i for i in ["1","0"]],columns=[i for i in ["Predict1","Predict0"]])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm,annot=True,fmt='g')
```
