```python
# data['Gender'].replace('NaN', np.NaN,inplace=True)
# data['Married'].replace('NaN', np.NaN,inplace=True)
data['Dependents'].replace('3+', 4,inplace=True)
# data['Dependents'].replace('NaN', np.NaN,inplace=True)
# data['Self_Employed'].replace('NaN', np.NaN,inplace=True)
```

```python
data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])
data['Married'] = data['Married'].fillna(data['Married'].mode()[0])
data['Self_Employed'] = data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])

data['Dependents'] = data['Dependents'].fillna(data['Dependents'].mode()[0])
data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].mode()[0])
```

```python
data['LoanAmount'] = data['LoanAmount'].fillna(data['LoanAmount'].median())
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].median())
```

```python
#to get dummies of categorical values
dummies=pd.get_dummies(data[['Loan_ID','Gender','Married','Dependents','Education','Self_Employed','Property_Area']],
                      columns=['Loan_ID','Gender','Married','Dependents','Education','Self_Employed','Property_Area'])
```

```python
dummies.head()
```

```python
data=pd.concat([data,dummies],axis=1)
```

```python
columns=['Loan_ID','Gender','Married','Dependents','Education','Self_Employed','Property_Area']
data.drop(columns,axis=1,inplace=True)
```

## Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
#linear model building
#independent variable and dependent variable
x=data.drop(['Loan_Status'],axis=1)
y=data[['Loan_Status']]
#split x and y data into 70:30 ratio
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=2)
#fit linear model
reg_model=LogisticRegression()
reg_model.fit(x_train,y_train)
```

```python

print(reg_model.score(x_train,y_train))
print(reg_model.score(x_test,y_test))
```

## DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
Dt=DecisionTreeClassifier(max_depth=3)
Dt.fit(x_train,y_train)
print(Dt.score(x_train,y_train))
print(Dt.score(x_test,y_test))
```

## BaggingClassifier

```python
from sklearn.ensemble import BaggingClassifier
Bc=BaggingClassifier(n_estimators=22,base_estimator=Dt,max_samples=20)
Bc.fit(x_train,y_train)
print(Bc.score(x_train,y_train))
print(Bc.score(x_test,y_test))
```

## AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier(n_estimators=2,random_state=1)
ad.fit(x_train,y_train)

print(ad.score(x_train,y_train))
print(ad.score(x_test,y_test))
```

## GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gc=GradientBoostingClassifier(n_estimators=7)
gc.fit(x_train,y_train)
print(gc.score(x_train,y_train))
print(gc.score(x_test,y_test))
```

## RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=400,random_state=1,max_samples=19)
rf.fit(x_train,y_train)
print(rf.score(x_train,y_train))
print(rf.score(x_test,y_test))
```
